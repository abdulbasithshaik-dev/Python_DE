```python
from pydantic import BaseModel
```

```python
class employee(BaseModel):
    name:str
    id:int
    department:str
```

```python
obj=employee(**{"name":"shaik","id":123,"department":"IT"})
obj
```

```python
from pydantic import StrictInt

class employee(BaseModel):
    name:str
    id:StrictInt
    department:str

obj=employee(**{"name":"shaik","id":123,"department":"IT"})
obj
```
